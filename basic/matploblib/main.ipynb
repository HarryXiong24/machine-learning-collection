# Matploblib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic

```python
np.random.seed(1)
x = range(20, 30)
y = np.random.randint(0, 10, 10)

lines = plt.plot(x, y, '--')

plt.title("Title") # title
plt.xlabel("x") # x label
plt.xticks(x, [f'{x} count' for x in range(20, 30)])# x ticks rotation
plt.grid(True) # grid 

line = lines[0]
line.set_linestyle('--')
line.set_linewidth(2)
```

## Line chart

```python
np.random.seed(1)
x = np.arange(20, 30)
y = np.random.randint(0, 10, 10)

plt.plot(x, y)
plt.show()

plt.plot(x, y, '--', color="#454fe4")
plt.show()

plt.plot(x, y, '--', color=(0, 0, 0, 0.5))
plt.show()
```

```python
np.random.seed(1)
meta_data = {
  'a':np.arange(20, 30),
  'b': np.random.randint(0, 10, 10)
}

plt.plot('a', 'b', '', data=meta_data, linewidth=2)
plt.show()
```
